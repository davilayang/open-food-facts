### How many columns have higher NA count?

```python
def count_na_pct(column):
    return df[column].isna().sum() / df.shape[0] * 100 
```

```python
count = 0; tmp = [] 
for col in df.columns:
    na_pct = count_na_pct(col)
    if na_pct > 50: 
        # print('NA Percent: ', count_na_pct(col), '\tAttribute: ', col)
        count += 1
        tmp.append(col)
print(count) # 143 columns have NA higher than 50%
```

```python
df2 = df.drop(tmp + ['states', 'states_tags', 'states_en'], axis=1)
df2.head()
```

```python
df2 = df[df['nutrition-score-fr_100g'].notna() | df['nutrition-score-uk_100g'].notna() ]
df2.shape
```

```python
def count_na_pct(column):
    return df2[column].isna().sum() / df2.shape[0] * 100 
```

```python
count = 0; tmp = [] 
for col in df2.columns:
    na_pct = count_na_pct(col)
    if na_pct > 80: 
        print('NA Percent: ', count_na_pct(col), '\tAttribute: ', col)
        count += 1
        tmp.append(col)
print(count) # 143 columns have NA higher than 50%
```
